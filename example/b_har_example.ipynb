# B-HAR

## Get started guide
In order to start with our analysis, we have to set up the B-HAR configuration file (see `config_example.cfg`).

One we are done with the configuration, we are ready to use B-HAR.

```python
from b_har.baseline import B_HAR
```

First of all, we need to crate a `B-HAR` object

```python
cfg_file = 'config_example.cfg'

b_har = B_HAR(config_file_path=cfg_file)
```

If we want to give a look to the dataset statistics we can run

```python
b_har.stats()
```

Start the framework and get the baseline for the input dataset

```python
b_har.get_baseline(['K-NN', 'DT', 'LDA'], ['m1_acc'])
```
